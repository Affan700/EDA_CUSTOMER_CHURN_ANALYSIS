# REPLCAED TENURE AS 0 , AND NO. OF TOTAL CHRGEES ARE RECORDED

# CONVERTED 0 AND 1 VALUES "SENIOR CITIZEN"

```python
def conv(value):
    if value ==1:
        return "yes"
    else :
        return "no"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(20)
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title ("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels = gb.index, autopct = "%1.2f%%")
plt.title ("Percentage of Churn Customers", fontsize = 10)
plt.show()
```

# FROM THE GIVEN PIE CHART WE CAN CONCLUDE THAT 26.54% OF OUR CUSTOMERS HAVE CHURNED OUT 
# NOW LETS EXPLORE THE REASON BEHIND IT.

```python
plt.figure(figsize = (3,4))
sns.countplot (x ="gender", data = df , hue = 'Churn')
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (3,4))
ax = sns.countplot (x ="SeniorCitizen", data = df )
ax.bar_label(ax.containers[0])
plt.title("Senior Citizen")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a crosstab to get percentages
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])
ct_pct = ct.apply(lambda x: x/x.sum() * 100, axis=1)

# Create the stacked bar chart
plt.figure(figsize=(8, 6))
ct_pct.plot(kind='bar', stacked=True)

# Add percentage labels on each segment
for i, (name, row) in enumerate(ct_pct.iterrows()):
    prev = 0
    for j, val in enumerate(row):
        plt.text(i, prev + val/2, f'{val:.1f}%', 
                 ha='center', va='center',
                 fontweight='bold', color='white')
        prev += val

# Customize the plot
plt.title("Churn Rate by Senior Citizen Status", fontsize=14)
plt.xlabel("Senior Citizen", fontsize=12)
plt.ylabel("Percentage (%)", fontsize=12)
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.legend(title='Churn')
plt.ylim(0, 100)

plt.tight_layout()
plt.show()
```

Overall, value-added support and security services appear to reduce churn more effectively than basic connectivity or streaming services.
